- AdaBoost: Adaptive boosting, or AdaBoost, starts by training a weak learner on the entire training set. It then evaluates it and assigns it a performance score depending on the model’s error. Next, it trains additional learners on the same dataset but with increased weights for the instances that were misclassified (or suffered from large prediction errors) by the previous learners. Finally, to make a prediction, AdaBoost combines the weak learners with each model’s vote weighted by its performance score.

- Gradient Boosting: This method is a generalization of boosting to arbitrary differentiable loss functions. Instead of tweaking sample weights like AdaBoost, gradient boosting fits each new learner to the residual errors (differences between the predicted and actual values) of the previous learner. You can think of it as a way of “correcting” the previous learners’ mistakes in the direction that minimally reduces the overall error. Eventually, the predictions of all learners are added up to make a final prediction.

- XGBoost: Extreme gradient boosting, or XGBoost, is a highly optimized implementation of gradient boosting that is designed to be more efficient and flexible than the standard version. While the steps for training a model are the same as for gradient boosting, it includes additional features such as regularization parameters to prevent overfitting, the ability to handle missing values, and an algorithm that helps to effectively find optimal splits in the data.

#### Model Requirement:
- Boosting: Use AdaBoost, Gradient Boosting, XGBoost
- Stacking: Layer 1: Two models, Layer 2: One model

Summary:
 
1️⃣ Boosting: Multiple models are trained in sequence, where each model attempts to correct the mistakes of the previous model.

2️⃣ Stacking: Multiple models are trained independently. We then train a meta-model using the outputs of these models as predictors.

```python
# AdaBoost - Increases the weight of incorrectly classified data and runs training again
from sklearn.ensemble import AdaBoostClassifier

clf1 = AdaBoostClassifier(n_estimators=100)

# Cross validation score
scores = cross_val_score(clf1, X, y, cv=5)
print("Cross Val Score:", scores.mean())

# Train the model
clf1.fit(X_train, y_train)

# Accuracy scores
print("Train Accuracy:", clf1.score(X_train, y_train))
print("Test Accuracy:", clf1.score(X_test, y_test))

# Compute the AUC score
auc_score = roc_auc_score(y_test, clf1.predict_proba(X_test)[:, 1])
print(f"AUC Score: {auc_score}")
```

```python
# Plot learning curves for AdaBoost
plot_learning_curves(X_train, y_train, X_test, y_test, clf1, print_model = True, style = 'ggplot')
plt.show()
```

```python
# Gradient Boosting - Trains a new model on residual errors made by the previous predictor
from sklearn.ensemble import GradientBoostingClassifier

clf2 = GradientBoostingClassifier(n_estimators=100, 
                                 learning_rate=1.0, 
                                 max_depth=1, 
                                 random_state=0)

# Cross validation score
scores = cross_val_score(clf2, X, y, cv=5)
print("Cross Val Score:", scores.mean())

# Train the model
clf2.fit(X_train, y_train)

# Accuracy scores
print("Train Accuracy:", clf2.score(X_train, y_train))
print("Test Accuracy:", clf2.score(X_test, y_test))

# Compute the AUC score
auc_score = roc_auc_score(y_test, clf2.predict_proba(X_test)[:, 1])
print(f"AUC Score: {auc_score}")
```

```python
# Plot learning curves for Gradient Boosting
plot_learning_curves(X_train, y_train, X_test, y_test, clf2, print_model = True, style = 'ggplot')
plt.show()
```

```python
# XGBoost - Specializes in residual gradient boosting techniques
# Model type 1, using XGBClassifier (Scikit-Learn API) - Suitable for simplified training. Automatically creates DMatrix
from xgboost import XGBClassifier

# Define the parameters for XGBoost
xgb_params = {
    'objective': 'binary:logistic', # Specifies the learning task as binary classification
    'max_depth': 3, # The maximum depth of each tree in the ensemble
    'learning_rate': 0.1, # The learning rate, controls the contribution of each tree
    'n_estimators': 100 # The number of boosting rounds or trees in the ensemble
}

# Create XGBClassifier
clf3 = XGBClassifier(**xgb_params)

# Cross validation score
scores = cross_val_score(clf3, X, y, cv=5)
print("Cross Val Score:", scores.mean())

# Train the model
clf3.fit(X_train, y_train)

# Accuracy scores
print("Train Accuracy:", clf3.score(X_train, y_train))
print("Test Accuracy:", clf3.score(X_test, y_test)) 

# Compute the AUC score
auc_score = roc_auc_score(y_test, clf3.predict_proba(X_test)[:, 1])
print(f"AUC Score: {auc_score}")
```

```python
# Plot learning curves for Gradient Boosting
plot_learning_curves(X_train, y_train, X_test, y_test, clf3, print_model = True, style = 'ggplot')
plt.show()
```

```python
# XGBoost - Specializes in residual gradient boosting techniques
# Model type 2, using xgb.train (Low-level API) - Suitable for maximum control and customization. Must define DMatrix
import xgboost as xgb

# Use DMatrix for XGBoost - Helps to improve training speed and memory efficiency
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=dv.feature_names_)
dtest = xgb.DMatrix(X_test, label=y_test, feature_names=dv.feature_names_)

# Set XGBoost params
xgb_params = {
    'eta': 0.3, # The training step for each iteration
   'max_depth': 5,  # The maximum depth of each tree
    'min_child_weight': 1, # Minimum sum of instance weight needed in a child
   'objective': 'binary:logistic',  # Specifies the learning task
    'nthread': 8, # Number of parallel threads
    'seed': 1, # Seed for random number generation
    'verbosity': 1  # Controls messages during training (0 for no messages, 1 for messages)
} 

clf4 = xgb.train(xgb_params, dtrain, num_boost_round=15)

# Make prediction
y_pred = clf4.predict(dtest)

# Convert predicted probabilities to binary predictions (0 or 1)
y_pred_binary = [1 if pred > 0.5 else 0 for pred in y_pred]

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_binary)
print("Test Accuracy:", accuracy)

# Calculate ROC AUC score
print("AUC Score:", roc_auc_score(y_test, y_pred))
```

```python
# Stacking Classifier - Results of individual predictions are treated as next training data
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression

# Using four models to improve metrics (instead of two as per project requirement)
estimators = [
    ("rf", RandomForestClassifier(n_estimators=100, random_state=42)),
    ("svc", make_pipeline(StandardScaler(), SVC(probability=True, random_state=42))),
    ("nb", GaussianNB()),
    ("knn", KNeighborsClassifier(n_neighbors=5))
]

# Choosing a robust final estimator, and tune its parameters if needed
final_estimator = LogisticRegression()

clf = StackingClassifier(estimators=estimators, final_estimator=final_estimator)

# Cross validation score
scores = cross_val_score(clf, X, y, cv=5)
print("Cross Val Score:", scores.mean())


# Train the model
clf.fit(X_train, y_train)

# Accuracy scores
print("Train Accuracy:", clf.score(X_train, y_train))
print("Test Accuracy:", clf.score(X_test, y_test)) 

# Compute the AUC score
auc_score = roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1])
print(f"AUC Score: {auc_score}")
```
