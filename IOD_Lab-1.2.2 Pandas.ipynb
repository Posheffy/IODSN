# 3. Make the index meaningful by making `EmployeeID` the index

````python
# A:
# df.set_index('x') is used to set the index. inplace=True means the changes will be reflected in the original DataFrame
# df.reset_index() is used to reverse df.set_index()
df.set_index('EmployeeID', inplace=True)
````

#### 3.A Find a specific row using the index

````python
# A:
# df.loc[] is used for label-based indexing, while df.iloc[] is used for integer position-based indexing
df.loc[1318]
# returns all the columns of data for employee 1318: label-based indexing
````

````python
# df.iloc[] is used for integer position-based indexing
df.iloc[0]
# returns data on first row, zero indexing [0]: position-based indexing
````

#### 3.B Sort the dataframe using the index

````python
# A:
# df.sort_index() is used to sort the index labels of a DataFrame
df.sort_index(ascending=False)
````

#### 3.C [BONUS] Sort the dataframe using `age`

#### 3.D [BONUS] Using `sample` select any row

# 4. Query Data Frame

#### 4.A Find rows Where `job_title` is CEO

#### 4.B Group by `age` and find how many there are of each age

# 5. Visualisation


#### 5.A Plot a histogram of `age`

````python
# A:
plt.hist(df['age'], bins=20, color='green', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
````

#### 5.B BoxPlot of `age`

````python
# A:
plt.boxplot(df['age'])
plt.grid(True, linestyle='--')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Boxplot of Age')
plt.show()
````

#### 5.C Create a scatterplot of `age` and `length_of_service`

````python
# A:
plt.scatter(df['age'], df['length_of_service'], c='purple', edgecolors='black')
plt.xlabel('Age')
plt.ylabel('Length of Service')
plt.title('Scatterplot of Age and Length of Service')
plt.show()
````
