````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# ANSWER - function to calculate m, b
# solve for m and b
def best_fit(X, Y):
    
    n = len(X)
    
    sum_x = sum(X)
    sum_y = sum(Y)
    sum_x_squared = sum(x**2 for x in X)
    sum_xy = sum(x*y for x, y in zip(X, Y))
    
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    b = (sum_y - m * sum_x) / n
    
    return m, b

# testing function and comparing with values of np.ployfit(x,y,1)
c, d = best_fit(x, y)
print(c, d)
````

````python
# ANSWER - plot of best fit line
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

#y = 0.80 + 0.92x

# solution
m, b = best_fit(X, Y)

# best fit line:
y = m * np.array(X) + b

# plot points and fit line
plt.scatter(X, Y)
plt.plot(X, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of best fit line')
plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

```
# Expected Output:
(30.476190476190474, 19.539561081583383, 41.41281987079756)
```

````python
# ANSWER
# function to calculate the mean and confidence level
from scipy import stats
def mean_confidence_interval(data, confidence=0.95): # default confidence interval is 95%
    a = 1 * np.array(data) # converts input data to Numpy array
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a) # mean and std error of mean
    h = se * scipy.stats.t.ppf((1 + confidence)/ 2, n-1) # margin of error for the confidence interval
    return m, m-h, m+h # sample mean, lower and higher bounds of the confidence interval

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))
````
