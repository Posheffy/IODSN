```python
# Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt
# Create a data set
x = np.linspace(0, 10, 100)
# Print the data
#print(x)
# Create a figure using plt.figure
fig = plt.figure(num=1, figsize=(8,6), facecolor='pink')
# Plot the sine wave of x
plt.plot(x, np.sin(x), '-')
# Plot the cosine wave of x
plt.plot(x, np.cos(x), '.')
```

```python
# Create two arrays of 5 arbitrary numbers
array1 = [2, 7, 1, 9, 4]
array2 = [5, 3, 8, 6, 0]
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "Example two"

plt.bar(np.arange(2*len(array1))[::2], array1, label="Example one", color='y', width=0.4)
plt.bar(np.arange(2*len(array1))[::2]+1, array2, label="Example two", color='r', width=0.4)

# Plot the legend
plt.legend()
# Label x axis "Bar number"
plt.xlabel('Bar number')
# Label y axis "Bar height"
plt.ylabel('Bar height')
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code
# Create an array of a random normal distribution
b = np.random.laplace(loc=15, scale=3, size=100)
# Create a histrogram using plt.hist()
n, bins, patches = plt.hist(b, bins='auto', color='#0504aa', alpha=0.75, rwidth=0.85)
# Set grid for x and y
# Set transparency of the grid using alpha=0.5
plt.grid(axis='both', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(25, 35, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```

```python
# Bonus: Create a scatter plot of random data
import matplotlib.pyplot as plt
import numpy as np

# Set the seed for reproducibility
np.random.seed(42)
num_point = 100
x = np.random.rand(num_point)
y = np.random.rand(num_point)

plt.figure(figsize = (8,6))
plt.scatter(x, y, color='blue', label='random data', marker='*')
plt.show()
```
